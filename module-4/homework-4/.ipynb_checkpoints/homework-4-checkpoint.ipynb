```python
import numpy
from matplotlib import pyplot
import matplotlib.cm as cm
%matplotlib inline
```

```python
n = 192

Du, Dv, F, k = 0.00016, 0.00008, 0.035, 0.065 # Bacteria 1

dh = 5/(n-1)

T = 8000

dt = .9 * dh**2 / (4*max(Du,Dv))

nt = int(T/dt)
```

```python
uvinitial = numpy.load('uvinitial.npz')
U_init = uvinitial['U']
V_init = uvinitial['V']
```

```python
fig = pyplot.figure(figsize=(8,5))
pyplot.subplot(121)
pyplot.imshow(U_init, cmap=cm.RdBu)
pyplot.xticks([]), pyplot.yticks([]);
pyplot.subplot(122)
pyplot.imshow(V_init, cmap=cm.RdBu)
pyplot.xticks([]), pyplot.yticks([]);
```

```python
def ftcs(U, V, Du, Dv, dt, dh, nt, F, k):
    
    Ut = numpy.zeros((nt,U.shape[0],U.shape[1]))
    
    for n in range(nt):
        Un = U.copy()
        Vn = V.copy()
        
        U[1:-1,1:-1] = Un[1:-1,1:-1] + (Du*dt/dh**2) *\
                       (Un[2:,1:-1] - 2*Un[1:-1,1:-1] + Un[:-2,1:-1] +\
                        Un[1:-1,2:] - 2*Un[1:-1,1:-1] + Un[1:-1,:-2]) -\
                       dt*Un[1:-1,1:-1]*Vn[1:-1,1:-1]**2 +\
                       dt*F*(1-Un[1:-1,1:-1])
        
        V[1:-1,1:-1] = Vn[1:-1,1:-1] + (Dv*dt/dh**2) *\
                       (Vn[2:,1:-1] - 2*Vn[1:-1,1:-1] + Vn[:-2,1:-1] +\
                        Vn[1:-1,2:] - 2*Vn[1:-1,1:-1] + Vn[1:-1,:-2]) +\
                       dt*Un[1:-1,1:-1]*Vn[1:-1,1:-1]**2 -\
                       dt*(F+k)*Vn[1:-1,1:-1]
        
        U[0,:] = U[1,:]
        U[-1,:] = U[-2,:]
        U[:,0] = U[:,1]
        U[:,-1] = U[:,-2]
        
        V[0,:] = V[1,:]
        V[-1,:] = V[-2,:]
        V[:,0] = V[:,1]
        V[:,-1] = V[:,-2]
        
        Ut[n] = U.copy()
    
    return Ut
```

```python
Ut = ftcs(U_init.copy(), V_init.copy(), Du, Dv, dt, dh, nt, F, k)
```

```python
fig = pyplot.figure(figsize=(8,6))
pyplot.xticks([]), pyplot.yticks([]);

im = pyplot.imshow(U_init, cmap=cm.RdBu)
```

```python
def init():
    im.set_data(U_init)
    return im,
```

```python
def animate(data):
    im.set_array(data)
    return im,
```

```python
from matplotlib import animation
from IPython.display import HTML
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=Ut, interval=1)
```

```python
HTML(anim.to_html5_video())
```
